```python
import matplotlib.pyplot as plt
```

```python
x = list(range(1,101))
y = []
y_win = []
```

```python
with open('stats.txt','r') as file:
    for _ in range(100):
        ln = file.readline()
        ln = ln.strip()
        y.append(int(ln))
        ln = file.readline()
        ln = ln.strip()
        y_win.append(int(ln)/500)
```

```python
plt.plot(x,y);
```

```python
plt.plot(x,y_win);
```
