```python
import matplotlib.pyplot as plt
```

```python
splits = 2000
x = list(range(1,splits+1))
y = []
y_win = []
```

```python
with open('stats.txt','r') as file:
    for _ in range(splits):
        ln = file.readline()
        ln = ln.strip()
        y.append(int(ln)/500)
        ln = file.readline()
        ln = ln.strip()
        y_win.append(int(ln)/500)
```

```python
plt.plot(x,y)
plt.xlabel(f'Split number (1-{splits})')
plt.ylabel("Average score")
plt.title(f'Average score of {splits} splits');
```

```python
plt.plot(x,y_win);
plt.xlabel(f'Split number (1-{splits})')
plt.ylabel("Win rate")
plt.title(f'Win rate of {splits} splits');
```
